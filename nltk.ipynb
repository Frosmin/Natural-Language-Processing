```python
import nltk 
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')


from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
```

```python
ejemplo = "HOla mi nombre es juan cada a a a dia me gusta ir a pasear a a a a a sere"
ejemplo_lower = ejemplo.lower()
print(ejemplo_lower)
```

```python
tokens = nltk.tokenize.word_tokenize(ejemplo_lower)

print(tokens)
```

```python

#stop word son palabras que son como a los les que no sirven xd
stop_words_espanol = set(stopwords.words('spanish'))
print(stop_words_espanol)
```

```python
puntuacion = list(string.punctuation)
print(puntuacion)
#signos de puntuacion mas extraños como &#!
```

```python

filtro= stop_words_espanol.union(puntuacion)
print(filtro)
```

```python
tokens_limpios  = []
for token in tokens:
    if token not in filtro:
        tokens_limpios.append(token)

print(tokens_limpios)
```
