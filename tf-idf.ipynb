```python
def obtener_raiz_casera(palabra):
    if palabra.endswith("ar"):
        return palabra[:-2]
    
    if palabra.endswith("er"):
        return palabra[:-2]
    
    if palabra.endswith("ir"):
        return palabra[:-2]
    
    if palabra.endswith("as"):
        return palabra[:-2]
    
    if palabra.endswith("es"):
        return palabra[:-2]
    
    if palabra.endswith("os"):
        return palabra[:-2]
    
    if palabra.endswith("s"):
        return palabra[:-1]

    
    
    return palabra

print(obtener_raiz_casera("niñas"))
```

```python
def procesar_texto_desde_cero(texto):
    texto = texto.lower()
    

    for signo in puntuacion:
        texto = texto.replace(signo, "")
        
    
    palabras = texto.split()
    
    # Stemming
    palabras_procesadas = []
    for palabra in palabras:
        if palabra not in stop_words_simon:
            raiz = obtener_raiz_casera(palabra)
            palabras_procesadas.append(raiz)
            
    
    return " ".join(palabras_procesadas)

print(procesar_texto_desde_cero("HOla, , amigos como estan habia muchas niñas por ahi"))
```

```python
lista_solo_mesajes = []
categoria = []

for dato  in datos_entrenamiento_procesados:
    lista_solo_mesajes.append(dato[0]) 
    categoria.append(dato[1])

print(lista_solo_mesajes)
```

```python


vocabulario  = set()
for mensaje in lista_solo_mesajes:
    palbras = mensaje.split()
    vocabulario.update(palbras)
vocabulario = sorted(list(vocabulario))



for i, palabra in enumerate(vocabulario):
    print(f"{i}: {palabra}")



print(len(vocabulario))
```

```python
N = len(lista_solo_mesajes)


def idf(lista_solo_mesajes,vocabulario):
    idf_diccionario  = {}


    for palabra in vocabulario:
        df = 0
        for mensaje in lista_solo_mesajes:
            palbras_del_nesaje = mensaje.split()

            if palabra in palbras_del_nesaje:
                df+=1
        
        idf_diccionario[palabra] = math.log((N + 1) / (df + 1)) + 1 #repasar bien xd
    
    return idf_diccionario



ejemplo = idf(lista_solo_mesajes,vocabulario)
print(ejemplo)
for palabra in ejemplo:
    print(f"{palabra} = {ejemplo[palabra]}")


```

```python
def tf(mensaje, vocabulario):
    palabras_mensaje = mensaje.split()
    vector_tf = []
    
    for palabra in vocabulario:
        conteo = palabras_mensaje.count(palabra)
        vector_tf.append(conteo)
        
    return vector_tf




ejemplo = tf(lista_solo_mesajes[0],vocabulario)
for i in range(len(vocabulario)):
    print(f"{vocabulario[i]} -> {ejemplo[i]}")
```

```python



def TFIDF(lista_solo_mensajes, vocabulario, idf_entrenado):
    matriz_tfidf = []


    for mensaje in lista_solo_mensajes:
        vector_tf = tf(mensaje, vocabulario)
        
        vector_tfidf = []
        for i, valor_tf in enumerate(vector_tf):
            palabra_actual = vocabulario[i]
            valor_idf = idf_entrenado.get(palabra_actual, 0)
            
            vector_tfidf.append(valor_tf * valor_idf)
        
        matriz_tfidf.append(vector_tfidf)
    return matriz_tfidf

idf_entrenado = idf(lista_solo_mesajes, vocabulario)

x = TFIDF(lista_solo_mesajes, vocabulario, idf_entrenado)



for item in x:
    print(item)
print(len(x))

print(len(lista_solo_mesajes[0]))
print(len(x[0]))

```

```python
modelo = MultinomialNB()
modelo.fit(x,categoria)
```
