```python
import nltk
import spacy
import string

# --- 1. Definir el Texto de Entrada y Filtros ---
texto_ejemplo = "HOla mi nombre es juan cada a a a dia me gusta ir a pasear a a a a a sere estuve"

# Crear el filtro (Stop Words + Puntuación)
stop_words_espanol = set(nltk.corpus.stopwords.words('spanish'))


puntuacion = set(string.punctuation)
# Usamos .union() para crear el filtro eficiente (SET)
filtro = stop_words_espanol.union(puntuacion)



```

```python
nlp = spacy.load('es_core_news_sm')
doc = nlp(texto_ejemplo.lower()) 
print(doc)
```

```python
lemas_limpios = [] 



for token in doc:
    lema = token.lemma_.lower()
    
    # Excluir espacios vacíos que spaCy a veces detecta
    if token.is_space:
        continue
    

    if lema not in filtro:
        lemas_limpios.append(lema)

print(lemas_limpios)
```
