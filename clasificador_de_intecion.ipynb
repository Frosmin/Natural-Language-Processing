```python

datos_entrenamiento = [
    # IntenciÃ³n: VENTA (Alta Prioridad)
    ("quiero comprar un auto sedan", "VENTA"),
    ("me interesa el precio de la camioneta 4x4", "VENTA"),
    ("busco financiamiento para un cero kilometros", "VENTA"),
    ("tienen disponible el modelo rojo?", "VENTA"),
    ("quiero agendar una prueba de manejo", "VENTA"),
    
    # IntenciÃ³n: SERVICIO (Media Prioridad)
    ("necesito cambio de aceite", "SERVICIO"),
    ("mi auto hace un ruido extraÃ±o", "SERVICIO"),
    ("cuÃ¡nto cuesta el mantenimiento de los 10mil km", "SERVICIO"),
    ("tienen repuestos para espejos?", "SERVICIO"),
    ("quiero agendar cita en el taller", "SERVICIO"),
    
    # IntenciÃ³n: INFO (Baja Prioridad)
    ("hola buenas tardes", "INFO"),
    ("a que hora abren?", "INFO"),
    ("donde estan ubicados?", "INFO"),
    ("tienen estacionamiento?", "INFO"),
    ("gracias", "INFO")
]

# Convertimos a DataFrame para manejo fÃ¡cil
df = pandas.DataFrame(datos_entrenamiento, columns=['mensaje', 'categoria'])

# --- PASO 2: CREAR EL PIPELINE (Vectorizador + Modelo) ---
# Usamos make_pipeline para unir el TF-IDF con el algoritmo Naive Bayes.
# El modelo aprenderÃ¡ automÃ¡ticamente quÃ© palabras pertenecen a quÃ© categorÃ­a.
modelo = make_pipeline(TfidfVectorizer(), MultinomialNB())

# --- PASO 3: ENTRENAR EL MODELO ---
# Le damos los mensajes (X) y sus categorÃ­as correctas (y)
modelo.fit(df['mensaje'], df['categoria'])

print("âœ… Modelo entrenado exitosamente.\n")
print("-" * 50)

```

```python
# --- PASO 4: SIMULACIÃ“N (Nuevos mensajes llegando por WhatsApp) ---
nuevos_mensajes = [
    "Mi auto exploto"       
]

print("CLASIFICACIÃ“N DE NUEVOS LEADS:\n")

def obtener_prioridad(categoria):
    if categoria == "VENTA":
        return "ðŸ”´ ALTA (Â¡Atender YA!)"
    elif categoria == "SERVICIO":
        return "ðŸŸ¡ MEDIA (Derivar a Taller)"
    else:
        return "ðŸŸ¢ BAJA (Bot o respuesta automÃ¡tica)"

# Predecir categorÃ­a para cada mensaje nuevo
predicciones = modelo.predict(nuevos_mensajes)

for mensaje, categoria_predicha in zip(nuevos_mensajes, predicciones):
    prioridad = obtener_prioridad(categoria_predicha)
    
    print(f"ðŸ“© Mensaje: '{mensaje}'")
    print(f"ðŸ“Š CategorÃ­a Detectada: {categoria_predicha}")
    print(f"ðŸ’¼ Prioridad de Negocio: {prioridad}")
    print("-" * 30)
```
