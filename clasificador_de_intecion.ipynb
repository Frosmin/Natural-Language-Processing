```python
def obtener_raiz_casera(palabra):
    if palabra.endswith("ar"):
        return palabra[:-2]

    if palabra.endswith("er"):
        return palabra[:-2]

    if palabra.endswith("ir"):
        return palabra[:-2]

    if palabra.endswith("as"):
        return palabra[:-2]

    if palabra.endswith("es"):
        return palabra[:-2]

    if palabra.endswith("os"):
        return palabra[:-2]

    if palabra.endswith("s"):
        return palabra[:-1]



    return palabra

print(obtener_raiz_casera("niñas"))
```

```python
from sklearn.metrics.pairwise import euclidean_distances, cosine_similarity
```

```python
if 'datos_entrenamiento_procesados' not in locals():
    datos_entrenamiento_procesados = [
        ["el precio es muy alto", "queja"],
        ["quiero comprar este producto", "compra"],
        ["hola buenos dias", "saludo"],
        ["donde queda la tienda", "consulta"]
    ]

lista_solo_mesaje = []
categoria = []

for dato in datos_entrenamiento_procesados:
    # Asumimos que el índice 0 es el texto y el 1 la etiqueta
    lista_solo_mesaje.append(dato[0])
    categoria.append(dato[1])

print(f"Total de mensajes procesados: {len(lista_solo_mesaje)}")
```

```python
#Usamos TF-IDF para dar peso a las palabras importantes y normalizar
vectorizer = TfidfVectorizer()

# Entrenamos el vectorizador con tus mensajes y los transformamos
X_train_vectors = vectorizer.fit_transform(lista_solo_mesaje)

print("Vectorización completada. Dimensiones de la matriz:", X_train_vectors.shape)
```

```python
def clasificar_mensaje(nuevo_mensaje, vectorizer, X_train, y_train):
    """
    Clasifica un nuevo mensaje comparándolo con la base de conocimientos
    usando 3 métricas diferentes.
    """
    # 1. Convertir el nuevo mensaje al mismo espacio vectorial
    nuevo_vec = vectorizer.transform([nuevo_mensaje])

    # --- A. Distancia Euclidiana ---
    # Calculamos la distancia contra TODOS los vectores de entrenamiento
    dists_euclidiana = euclidean_distances(nuevo_vec, X_train)
    # Buscamos el índice del valor MÍNIMO (menor distancia = más parecido)
    idx_eucl = dists_euclidiana.argmin()
    val_eucl = dists_euclidiana[0][idx_eucl]

    # --- B. Similitud del Coseno ---
    cos_sims = cosine_similarity(nuevo_vec, X_train)
    # Buscamos el índice del valor MÁXIMO (mayor similitud = más parecido)
    idx_cos_sim = cos_sims.argmax()
    val_cos_sim = cos_sims[0][idx_cos_sim]

    # --- C. Distancia del Coseno ---
    # Se puede derivar de la similitud: 1 - similitud
    # O calcular explícitamente usando pairwise_distances con metric='cosine'
    val_cos_dist = 1 - val_cos_sim
    # El índice es el mismo que el de similitud (si la sim es max, la dist es min)
    idx_cos_dist = idx_cos_sim

    # --- Resultados ---
    return {
        "mensaje_original": nuevo_mensaje,
        "euclidiana": {
            "categoria_predicha": y_train[idx_eucl],
            "mensaje_mas_cercano": lista_solo_mesaje[idx_eucl],
            "valor_distancia": round(val_eucl, 4) # Menor es mejor
        },
        "coseno_similitud": {
            "categoria_predicha": y_train[idx_cos_sim],
            "mensaje_mas_cercano": lista_solo_mesaje[idx_cos_sim],
            "valor_similitud": round(val_cos_sim, 4) # Mayor es mejor (max 1.0)
        },
        "coseno_distancia": {
            "categoria_predicha": y_train[idx_cos_dist],
            "mensaje_mas_cercano": lista_solo_mesaje[idx_cos_dist],
            "valor_distancia": round(val_cos_dist, 4) # Menor es mejor (min 0.0)
        }
    }
```

```python
# Mensaje nuevo que queremos clasificar
mensaje_prueba = "cuanto cuesta el producto"

resultados = clasificar_mensaje(mensaje_prueba, vectorizer, X_train_vectors, categoria)

print(f"--- Análisis para: '{mensaje_prueba}' ---\n")

print(f"1. Según Distancia Euclidiana:")
print(f"   -> Clasificado como: {resultados['euclidiana']['categoria_predicha']}")
print(f"   -> Basado en mensaje similar: '{resultados['euclidiana']['mensaje_mas_cercano']}'")
print(f"   -> Distancia: {resultados['euclidiana']['valor_distancia']} (Cerca de 0 es mejor)\n")

print(f"2. Según Similitud del Coseno:")
print(f"   -> Clasificado como: {resultados['coseno_similitud']['categoria_predicha']}")
print(f"   -> Basado en mensaje similar: '{resultados['coseno_similitud']['mensaje_mas_cercano']}'")
print(f"   -> Similitud: {resultados['coseno_similitud']['valor_similitud']} (Cerca de 1 es mejor)\n")

print(f"3. Según Distancia del Coseno:")
print(f"   -> Clasificado como: {resultados['coseno_distancia']['categoria_predicha']}")
print(f"   -> Distancia: {resultados['coseno_distancia']['valor_distancia']} (Cerca de 0 es mejor)")
```

```python
def imprimir_comparacion_modelos(mensaje, vectorizer, X_matrix, etiquetas):
    """
    Toma un mensaje, calcula sus clasificaciones usando Euclidian, Coseno Similitud
    y Coseno Distancia, e imprime un reporte comparativo legible.

    Args:
        mensaje (str): El texto nuevo a clasificar.
        vectorizer (obj): El objeto TfidfVectorizer ya entrenado.
        X_matrix (sparse matrix): La matriz de vectores de entrenamiento.
        etiquetas (list): La lista de categorías correspondientes a la matriz.
    """

    # 1. Obtenemos los cálculos matemáticos llamando a la función anterior
    resultados = clasificar_mensaje(mensaje, vectorizer, X_matrix, etiquetas)

    # 2. Imprimimos el reporte formateado
    print("=" * 60)
    print(f"--- ANÁLISIS PARA: '{mensaje}' ---")
    print("=" * 60)

    # Bloque A: Distancia Euclidiana
    res_eucl = resultados['euclidiana']
    print(f"\n1. [Distancia Euclidiana]")
    print(f"   • Clasificación:  {res_eucl['categoria_predicha'].upper()}")
    print(f"   • Mensaje vecino: '{res_eucl['mensaje_mas_cercano']}'")
    print(f"   • Distancia:      {res_eucl['valor_distancia']} (Cerca de 0 es mejor)")

    # Bloque B: Similitud del Coseno
    res_cos = resultados['coseno_similitud']
    print(f"\n2. [Similitud del Coseno]")
    print(f"   • Clasificación:  {res_cos['categoria_predicha'].upper()}")
    print(f"   • Mensaje vecino: '{res_cos['mensaje_mas_cercano']}'")
    print(f"   • Similitud:      {res_cos['valor_similitud']} (Cerca de 1 es mejor)")

    # Bloque C: Distancia del Coseno
    res_dist = resultados['coseno_distancia']
    print(f"\n3. [Distancia del Coseno]")
    print(f"   • Clasificación:  {res_dist['categoria_predicha'].upper()}")
    print(f"   • Distancia:      {res_dist['valor_distancia']} (Cerca de 0 es mejor)")
    print("-" * 60 + "\n")
```

```python
def entrenar_naive_bayes(vectorizer, X_train_vecs, y_train):
    """
    Entrena un modelo Multinomial Naive Bayes.
    Este modelo es ideal para conteos de palabras (TF o TF-IDF).
    """
    # Instanciamos el modelo
    clf_bayes = MultinomialNB()

    # Entrenamos con la matriz vectorizada y las etiquetas
    clf_bayes.fit(X_train_vecs, y_train)

    return clf_bayes

# Entrenamos el modelo usando los vectores que ya tenías (X_train_vectors)
modelo_bayes = entrenar_naive_bayes(vectorizer, X_train_vectors, categoria)

print("Modelo Naive Bayes entrenado exitosamente.")
```

```python
from collections import defaultdict
import math
```

```python
class SimpleMarkovClassifier:
    def __init__(self):
        # Almacena las probabilidades de transición por categoría
        # Estructura: self.transitions[categoria][palabra_actual][siguiente_palabra] = probabilidad
        self.transitions = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))
        self.class_counts = defaultdict(int)
        self.vocab = set()

    def train(self, textos, etiquetas):
        """Aprende las transiciones entre palabras para cada categoría."""
        for texto, etiqueta in zip(textos, etiquetas):
            self.class_counts[etiqueta] += 1
            words = texto.lower().split()
            # Añadimos un token de inicio y fin para capturar el contexto completo
            words = ['<START>'] + words + ['<END>']

            for i in range(len(words) - 1):
                curr_w, next_w = words[i], words[i+1]
                self.transitions[etiqueta][curr_w][next_w] += 1
                self.vocab.add(curr_w)
                self.vocab.add(next_w)

        # Convertir conteos a probabilidades (Suavizado simple)
        for cat in self.transitions:
            for curr_w in self.transitions[cat]:
                total = sum(self.transitions[cat][curr_w].values())
                for next_w in self.transitions[cat][curr_w]:
                    # Probabilidad P(next | curr, category)
                    self.transitions[cat][curr_w][next_w] /= total

    def predict_score(self, texto, category):
        """Calcula la probabilidad (log-likelihood) de una frase dada una categoría."""
        words = texto.lower().split()
        words = ['<START>'] + words + ['<END>']
        score = 0

        # Sumamos logaritmos para evitar underflow numérico
        for i in range(len(words) - 1):
            curr_w, next_w = words[i], words[i+1]
            prob = self.transitions[category][curr_w].get(next_w, 0.00001) # 0.00001 es un valor epsilon pequeño
            score += math.log(prob)

        return score

    def predict(self, texto):
        """Retorna la categoría con mayor probabilidad para el texto."""
        best_cat = None
        best_score = -float('inf')

        for cat in self.class_counts:
            score = self.predict_score(texto, cat)
            if score > best_score:
                best_score = score
                best_cat = cat
        return best_cat

# Entrenamos el modelo de Markov
modelo_markov = SimpleMarkovClassifier()
modelo_markov.train(lista_solo_mesaje, categoria)

print("Modelo de Markov (Custom) entrenado exitosamente.")
```

```python
def probar_modelos_probabilisticos(mensaje, vectorizer, modelo_nb, modelo_mk):

    # 1. Predicción con Naive Bayes
    # Transformamos el mensaje al formato que Bayes espera
    mensaje_vec = vectorizer.transform([mensaje])
    pred_bayes = modelo_nb.predict(mensaje_vec)[0]
    # Obtenemos las probabilidades crudas para ver la confianza
    probs_bayes = modelo_nb.predict_proba(mensaje_vec)
    confianza_bayes = np.max(probs_bayes)

    # 2. Predicción con Markov
    pred_markov = modelo_mk.predict(mensaje)

    # 3. Imprimir Resultados
    print("="*60)
    print(f"--- ANÁLISIS PROBABILÍSTICO: '{mensaje}' ---")
    print("="*60)

    print(f"\n1. [Naive Bayes Multinomial]")
    print(f"   • Clasificación: {pred_bayes.upper()}")
    print(f"   • Lógica: Basado en conteo de palabras independientes.")
    print(f"   • Confianza: {confianza_bayes:.4f}")

    print(f"\n2. [Cadenas de Markov (Secuencial)]")
    print(f"   • Clasificación: {pred_markov.upper()}")
    print(f"   • Lógica: Basado en la probabilidad de que la palabra B siga a la A.")
    print("-" * 60 + "\n")
```
