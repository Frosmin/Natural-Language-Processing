```python
def lematizar_texto(texto):
    texto_minusculas = texto.lower()
    doc = nlp(texto_minusculas)
    
    # 2. Extraer el lema de cada palabra, filtrando stop_words y puntuaci√≥n
    tokens_limpios = []
    for token in doc:
        if token.text not in stop_words_simon and token.text not in puntuacion:
            tokens_limpios.append(token.lemma_)
            
    # 3. Unir de nuevo en una frase
    return " ".join(tokens_limpios)
```
