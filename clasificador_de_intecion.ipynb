```python
def obtener_raiz_casera(palabra):
    if palabra.endswith("ar"):
        return palabra[:-2]
    
    if palabra.endswith("er"):
        return palabra[:-2]
    
    if palabra.endswith("ir"):
        return palabra[:-2]
    
    if palabra.endswith("as"):
        return palabra[:-2]
    
    if palabra.endswith("es"):
        return palabra[:-2]
    
    if palabra.endswith("os"):
        return palabra[:-2]
    
    if palabra.endswith("s"):
        return palabra[:-1]

    
    
    return palabra

print(obtener_raiz_casera("niñas"))
```

```python
def procesar_texto_desde_cero(texto):
    texto = texto.lower()
    

    for signo in puntuacion:
        texto = texto.replace(signo, "")
        
    
    palabras = texto.split()
    
    # Stemming
    palabras_procesadas = []
    for palabra in palabras:
        if palabra not in stop_words_simon:
            raiz = obtener_raiz_casera(palabra)
            palabras_procesadas.append(raiz)
            
    
    return " ".join(palabras_procesadas)

print(procesar_texto_desde_cero("HOla, , amigos como estan habia muchas niñas por ahi"))
```

```python

datos_entrenamiento = [
    # Intención: VENTA AUTO
    ("quiero comprar un auto sedan", "VENTA"),
    ("me interesa el precio de la camioneta 4x4", "VENTA"),
    ("busco financiamiento para un cero kilometros", "VENTA"),
    ("tienen disponible el modelo rojo?", "VENTA"),
    ("quiero agendar una prueba de manejo", "VENTA"),
    ("quiero compara otro auto", "VENTA"),
    
    #VENTA PARTES
    ("quiero compara partes", "VENTA_PARTES"),
    ("hola tiene este repuesto?", "VENTA_PARTES"),
    ("hola tienen respuestos para toyota?", "VENTA_PARTES"),
    ("repuestos", "VENTA_PARTES"),



    # Intención: SERVICIO 
    ("necesito cambio de aceite", "SERVICIO"),
    ("mi auto hace un ruido extraño", "SERVICIO"),
    ("cuánto cuesta el mantenimiento de los 10mil km", "SERVICIO"),
    ("tienen repuestos para espejos?", "SERVICIO"),
    ("quiero agendar cita en el taller", "SERVICIO"),
    ("mi auto se choco", "SERVICIO"),
    
    # Intención: INFO 
    ("hola buenas tardes", "INFO"),
    ("a que hora abren?", "INFO"),
    ("donde estan ubicados?", "INFO"),
    ("tienen estacionamiento?", "INFO"),
]
```

```python

datos_entrenamiento_procesados = []

for dato in datos_entrenamiento:
    datos_entrenamiento_procesados.append((procesar_texto_desde_cero(dato[0]),dato[1]))

for dato in datos_entrenamiento_procesados:
    print(dato)
```

```python
lista_solo_mesaje = []
categoria = []

for dato  in datos_entrenamiento_procesados:
    lista_solo_mesaje.append(dato[0]) 
    categoria.append(dato[1])
```

```python
vectorizador  = TfidfVectorizer()

x = vectorizador.fit_transform(lista_solo_mesaje)


print(x)
```

```python
modelo = MultinomialNB()
modelo.fit(x,categoria)
```

```python

mensaje  = "hola buenas tardes quiero comprar un repuesto"
masaje_limpio = procesar_texto_desde_cero(mensaje)
mensaje_vectorizado = vectorizador.transform([masaje_limpio])


predicion = modelo.predict(mensaje_vectorizado)[0]
print(predicion)
```
